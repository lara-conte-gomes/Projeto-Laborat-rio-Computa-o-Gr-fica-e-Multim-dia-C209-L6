1 - INVERTER A IMAGEM NO EIXO A CADA 20 SEGUNDOS

```python
import moviepy.editor as mpe
import numpy as np
import matplotlib.pyplot as plt

video = mpe.VideoFileClip("C:/Users/larac/OneDrive/Área de Trabalho/Trabalho Prático 2/WIN_20241130_15_55_44_Pro.mp4")

#Reflexão foi feita no eixo x
def reflect(frame):
    img_refl = np.zeros(shape=frame.shape, dtype=np.uint8)
    img_refl[::-1, :] = frame
    return img_refl


# Função para aplicar reflexão condicional a cada 20 segundos alternados
def reflect_conditionally(get_frame, t):
    if int(t) // 20 % 2 == 1:  # Refletir para períodos múltiplos de 20 segundos (20-40, 60-80, ...)
        return reflect(get_frame(t))
    else:
        return get_frame(t)  # Mantém o frame normal


video_reflec = video.fl_image(reflect)
```

2 - DIMINUIR O SOM, DE FORMA GRADATIVA, A CADA 30 SEGUNDOS, DE FORMA QUE OS ÚLTIMOS 10 SEGUNDOS FIQUE SEM SILÊNCIO (SEM SOM)

```python
def decrease(audio, t):
    # Garantir que t seja um array (tratando processamento em lote)
    t = np.atleast_1d(t)

    # Inicializar lista para armazenar frames processados
    processed_frames = []

    for ti in t:
        frame = audio(ti)
        time_left = video.duration - ti

        # Intervalos de 30 segundos: diminui 10% a cada intervalo
        interval = int(ti) // 30  # Conta quantos intervalos de 30 segundos passaram
        volume_factor = max(1 - interval * 0.3, 0)  # A cada intervalo de 30 segundos, diminui 10% do volume

        # Se o tempo restante for menor ou igual a 10 segundos, o volume será 0 (silêncio)
        if time_left <= 10:
            volume_factor = 0

        # Ajustar o volume do frame
        if len(frame.shape) == 1:  # Áudio mono
            frame *= volume_factor
        else:  # Áudio estéreo
            frame[:, 0] *= volume_factor
            frame[:, 1] *= volume_factor

        processed_frames.append(frame)

    return np.array(processed_frames) if len(processed_frames) > 1 else processed_frames[0]

# Aplicar reflexão condicional ao vídeo
video_reflec = video.fl(reflect_conditionally)

# Aplicar redução de áudio
audio_decrease = video_reflec.audio.fl(decrease)
video_audio_decrease = video_reflec.set_audio(audio_decrease)
```

3 - APÓS O PRIMEIRO MINUTO, 60 SEGUNDOS, FAZER UM CORTE PARA O SEGUNDO 80

```python
part_1 = video_audio_decrease.subclip(0, 60)  # Do início até 60 segundos
part_cut = video_audio_decrease.subclip(60, 80) # Trecho cortado (60 a 80 segundos)
part_2 = video_audio_decrease.subclip(80)    # De 80 segundos até o final
```

4 - POR FIM, INSIRA OS 20 SEGUNDOS CORTADOS ACIMA NO FIM DO VÍDEO

```python
final_video = mpe.concatenate_videoclips([part_1, part_2, part_cut]) #Fazendo a junção da parte 1 (início até 60 segundos) com a parte 2 (80 segundos até o final) e com a part_cut (60 segundos a 80 segundos)
```

5 - SALVAR O VÍDEO NA PASTA DO TRABALHO

```python
# Exportar o vídeo final
final_video.write_videofile("resultado5_trabalho.mp4", codec="libx264", fps=24)
```
